# RETURN BANNER TO CHECK FOR SECURITY VULNERABILITY

```python
import socket

def retBanner(ip, port):
    try:
        socket.setdefaulttimeout(2)
        sock = socket.socket()
        sock.connect((ip,port))
        banner = sock.recv(1024)
        return banner.decode()
        
    except:
        return

def main(ip, ports):
    for port in ports:
        banner = retBanner(ip, port)
        if banner:
            out = '\n\t'.join([line.strip() for line in banner.split('\n')])
            print(f"[+] {ip} / {port} :\n\t{out}")
main(ip, ports)
```
