```python
import pyaudio
import struct
import numpy as np
import matplotlib.pyplot as plt
import time

%matplotlib tk
#inline
```

```python
chunk = 4000 #1024 * 4
Format = pyaudio.paInt16
channels = 1
rate = 16000 #44100 # samples per second
```

```python
p = pyaudio.PyAudio()
stream = p.open(
    format=Format,
    channels=channels,
    rate = rate,
    input = True,
    output = True,
    frames_per_buffer=chunk
)

data = stream.read(chunk)
print(len(data))
print(data)
```

```python
data_int = np.array(struct.unpack(str(2*chunk) + 'B', data), dtype='b')[::4]+127 # see the len(data) it is double the size of chunk
data_int
```

```python
plt.figure(figsize=(20,4))
#plt.ion()
plt.plot(data_int,'-', c='g')
plt.show()
```

```python
p = pyaudio.PyAudio()
stream = p.open(
    format=Format,
    channels=channels,
    rate = rate,
    input = True,
    output = True,
    frames_per_buffer=chunk
)


fig, ax = plt.subplots()
x = np.arange(0,2*chunk, 2)
line, = ax.plot(x,np.random.rand(chunk)) # single chunk becz we are slicing it later in down V

while True:
    data = stream.read(chunk)
    data_int = np.array(struct.unpack(str(2*chunk)+'B',data),dtype='b')[::2]+127#       here
    line.set_ydata(data_int)
    fig.canvas.draw()
    fig.canvas.flush_events()
```

```python
def Decimal2Binary(decimal):
    l = []
    reminder, temp = str(decimal%2), decimal//2
    l.append(reminder)
    while temp != 0:
        reminder, temp = str(temp%2), temp//2
        l.append(reminder)
    binary  = ''.join(l[::-1])
    return binary
```
